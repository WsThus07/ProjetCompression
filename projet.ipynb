```python
from tkinter import *
from PIL import ImageTk, Image
import os
from tkinter import messagebox
from tkinter import filedialog
import matplotlib, sys
matplotlib.use('TkAgg')
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib import pylab as plt
from scipy import ndimage
def show():
    global ltxt
    #top=Toplevel(window_main)
    text=ltxt.cget("text")
    if len(text)==0:
        messagebox.showerror("Error","File Unchosen")
        #print(type(text),text)
    else:
        ima= Image.open(text)
        ima.show()
def label_Lc():
    global ltxt,label_sizec,label_sizecF
    import IRM
    textc=ltxt.cget("text")
    initiale=IRM.ORG_file(textc)
    print(initiale)
    #nom,ext=textc.split('.')
    #file=nom+'ORG.'+ext
    finale=IRM.compression_file_irm(textc)
    print(finale)
    label_sizecI.config(text='taille initiale : '+str(round(initiale / 1024, 3)) + ' KB')
    label_sizecF.config(text='taille compressée : '+str(round(finale / 1024, 3)) + ' KB')
    t=IRM.taux_compression(initiale,finale)
    label_Taux.config(text='Taux de Compression : '+str(t) + ' %')
    
def plot():
    from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
    from matplotlib.figure import Figure
    import IRM
    global ltxt,label_sizec,label_sizecF
    textc=ltxt.cget("text")
    nom,ext=textc.split('.')
    path=nom+'.IRM'
    data=IRM.décompression(path)
    top=Toplevel(window_main)
    fig = plt.figure(figsize=(5,4))
    im = plt.imshow(data) # later use a.set_data(new_data)
    ax = plt.gca()
    ax.set_xticklabels([]) 
    ax.set_yticklabels([])

    # a tk.DrawingArea
    canvas = FigureCanvasTkAgg(fig, master=top)
    canvas.draw()
    canvas.get_tk_widget().pack(side=TOP, fill=BOTH, expand=1)

    
    
    

#start
def make_label():
        global im,label_file,label_size,ltxt
        foor.filename = filedialog.askopenfilename(initialdir = "/",title='select a file', filetypes =[('PNG Files', '*.png'),('IRM Files', '*.irm'),('TIFF Files', '*.tif'),('JPG Files', '*.jpg')])
        new_window=Toplevel(window_main)
        new_window.geometry('400x400')
        ima= Image.open(foor.filename)
        re = ima.resize((400, 400))
        i = ImageTk.PhotoImage(re)
        im=ImageTk.PhotoImage(Image.open(foor.filename))
        image_top=Label(new_window,image=im)
        image_top.pack()
        """im= Image.open(foor.filename)
        re = im.resize((300, 200))
        i = ImageTk.PhotoImage(re)"""
        #label_file.config(image=im)
        #size=os.path.getsize(foor.filename)
        #label_size.config(text='taille originale : '+str(round(size / 1024, 3)) + ' KB')
        ltxt.config(text=os.path.basename(foor.filename))
        # Resize the image using resize() method
        #resize_image2= imag.resize((400,400))

#end
window_main = Tk(className=' CoDe.irm')
window_main.geometry("900x800")
window_main.maxsize(width=900,height=800)
window_main.iconbitmap('irm.ico')
#Set window background color
window_main.config(background = "white")




p = PanedWindow(window_main, orient=HORIZONTAL)
p.pack(side=TOP, expand=Y, fill=BOTH, pady=2, padx=2)
one=Label(p, text='', background='black', anchor=N) 
p.add(one)
two=Label(p, text='', background='#630000', anchor=CENTER)
p.add(two)
three=Label(p, text='', background='grey', anchor=CENTER)
p.add(three)
foor=Label(p, text='    ', background='white', anchor=CENTER)
p.add(foor)
p.pack()
#
scrollbar = Scrollbar(foor, bg="blue")
scrollbar.pack( side = RIGHT, fill = Y )
#
image1_window_main = Image.open('backgd.jpg')
# Resize the image using resize() method
resize_image1_window_main = image1_window_main.resize((900, 800))
img_window_main = ImageTk.PhotoImage(resize_image1_window_main)
logo_window_main=Label(foor,image=img_window_main)
logo_window_main.place(x=0,y=0)


#
frame=Frame(p,width=900, height=100,bg = '#F7F7F7')
l=Label(frame,text='  ',width='300',bg='white')
image = Image.open('irm.png')
resize_image = image.resize((90, 90))
imgR = ImageTk.PhotoImage(resize_image)
logo=Label(foor,image=imgR).place(x=750,y=40)

#
imag_cadre= Image.open("cadre.png")
resize_cadre = imag_cadre.resize((500, 200))
cadre = ImageTk.PhotoImage(resize_cadre)
Le=Label(foor,image=cadre,width=500)
Le.place(x=60,y=40)
#
Buttontexte=Button(Le,text='choose an image',bg='white',width=30,command=make_label)
Buttontexte.place(x=80,y=30)
Button_show=Button(Le,text='show',bg='white',command=show)
Button_show.place(x=300,y=120)

#
ltxt=Label(Le,text='',width='30',bg='white')
ltxt.place(x=83,y=79)

#
#label_size=Label(Le,font='Candara',width="30",height="1",bg='white',bd=1)
#label_size.place(x=90,y=150)




#
DEFIM = Image.open('original2.png')
DERI = DEFIM.resize((100, 100))
default=ImageTk.PhotoImage(DERI)
label_file=Label(Le,image=default,bg='white',width="100",height="100")
label_file.place(x=355,y=27)
#
#partie compression
#
imag_cadrec= Image.open("cadre.png")
resize_cadrec = imag_cadrec.resize((500, 200))
cadrec = ImageTk.PhotoImage(resize_cadrec)
Lc=Label(foor,image=cadrec,width=500)
Lc.place(x=60,y=240)
#
Buttontextec=Button(Lc,text='compress',bg='white',width=30,command=label_Lc)
Buttontextec.place(x=80,y=30)
'''Button_showc=Button(Lc,text='show',bg='white')
Button_showc.place(x=355,y=60)'''
#
ltxtc=Label(Lc,text='',width='30',bg='white')
ltxtc.place(x=83,y=79)


#
label_sizecI=Label(Lc,font='Candara',width="30",height="1",bg='white',bd=1)
label_sizecI.place(x=10,y=60)
label_sizecF=Label(Lc,font='Candara',width="30",height="1",bg='white',bd=1)
label_sizecF.place(x=10,y=110)
label_Taux=Label(Lc,width='40',bg='white',bd=1)
label_Taux.place(x=45,y=160)#taux_de compression

#
DEFIMc = Image.open('compressed.png')
DERIc = DEFIMc.resize((100, 100))
defaultc=ImageTk.PhotoImage(DERIc)
label_filec=Label(Lc,image=defaultc,bg='white',width="100",height="100")
label_filec.place(x=355,y=27)

#partie decompression
#
#
imag_cadred= Image.open("cadre.png")
resize_cadred = imag_cadred.resize((500, 200))
cadred = ImageTk.PhotoImage(resize_cadred)
Ld=Label(foor,image=cadred,width=500)
Ld.place(x=60,y=440)
#
Buttontexted=Button(Ld,text='decompress',bg='white',width=30,command=plot)
Buttontexted.place(x=80,y=50)
Button_showd=Button(Ld,text='show',bg='white',command=show)
Button_showd.place(x=300,y=120)
#
ltxtd=Label(Ld,text='',width='30',bg='white')
ltxtd.place(x=83,y=79)
#
label_sized=Label(Ld,font='Candara',width="30",height="1",bg='white',bd=1)
label_sized.place(x=90,y=150)
#
DEFIMd = Image.open('decompressed2.png')
DERId = DEFIMd.resize((100, 100))
defaultd=ImageTk.PhotoImage(DERId)
label_filed=Label(Ld,image=defaultd,bg='white',width="100",height="100")
label_filed.place(x=355,y=27)
#

#
imag = Image.open("exit.png")
resize_imag = imag.resize((30, 20))
photo = ImageTk.PhotoImage(resize_imag)
button_exit = Button(window_main,text = "Exit",image=photo,width=30,command = window_main.destroy).place(x=820,y=566)
#
l.pack()
frame.pack(side=TOP)
window_main.mainloop()
```
